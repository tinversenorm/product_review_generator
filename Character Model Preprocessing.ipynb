```python
import numpy as np
import pandas as pd
import string
import re
import itertools
import io
```

```python
dataset1=pd.read_csv("1429_1.csv")
dataset1.head()
```

```python
def clean1(string1):
    string1=str(string1)
    punctuation1=string.punctuation
    punctuation1=punctuation1.replace(".","")
    punctuation1=punctuation1.replace(",","")
    punctuation1=punctuation1.replace("!","")
    punctuation1=punctuation1.replace("?","")
    punctuation1=punctuation1.replace('\'',"")
    #string1=dataset["reviews.txt"]
    string1=string1.translate(string1.maketrans("","",punctuation1))
    return string1
```

```python
dataset1["review.text.cleaned"]=dataset1["reviews.text"].apply(clean1)
dataset1["review.text.cleaned"].head()
```

```python
#sending the word arrays to a .out file
charTxt1 = dataset1["review.text.cleaned"].values
charTxt1 = " ".join(charTxt1)
```

```python
#sending the words arrays to a .txt file
outFile= open("AmazonChar.txt","w+",encoding='utf8')
outFile.write(charTxt1)
outFile.close()
```
