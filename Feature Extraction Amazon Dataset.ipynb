```python
import numpy as np
import pandas as pd
import string
import re
import itertools
import io
```

```python
dataset=pd.read_csv("1429_1.csv")
dataset.head()
```

```python
dataset["reviews.text"].head()
```

```python
#remove punctuation

#find something that only allows upper case, lower case, period, apostrophe, comma and put it all into a new column
#column will be called "review.text.cleaned"
```

```python
def clean1(string1):
    string1=str(string1)
    punctuation1=string.punctuation
    punctuation1=punctuation1.replace(".","")
    punctuation1=punctuation1.replace(",","")
    punctuation1=punctuation1.replace('\'',"")
    #string1=dataset["reviews.txt"]
    string1=string1.translate(string1.maketrans("","",punctuation1))
    return string1
```

```python
dataset["review.text.cleaned"]=dataset["reviews.text"].apply(clean1)
dataset["review.text.cleaned"].head()
```

```python
#put all the words into an array

#create new column that takes the text, removes ALL punctuation (only upper and lower case letters and apostrophe allowed)
# and then splits based on space (" ") add the array into a new column
# new column will be named "text.array"
```

```python
#new version
def wordArray(string2):
    string2=str(string2)
    punctuation2=string.punctuation
    punctuation2=punctuation2.replace('\'',"")
    punctuation2=punctuation2.replace(',',"")
    punctuation2=punctuation2.replace('.',"")
    punctuation2=punctuation2.replace('!',"")
    punctuation2=punctuation2.replace('?',"")
    returnString=re.sub(r'([,.!?])', r' \1', string2)
    returnString=returnString.translate(returnString.maketrans("","",punctuation2))
    returnString=returnString.replace("  ", " ")
    arr=returnString.split(" ")
    arr.append("<eor>")
    return (arr)
```

```python
dataset["review.words"]=dataset["reviews.text"].apply(wordArray)
dataset["review.words"].head()
```

```python
print(dataset["review.words"][5])
```

```python
#add in the <endofreview> column

#get the last word of the array and put it in a new column
```

```python
def lastWord(string3):
    string3=str(string3)
    string3=string3.translate(string3.maketrans("","",string.punctuation))
    arr=string3.split(" ")
    size=len(arr)
    return arr[size-1]
```

```python
dataset["endofreview"]=dataset["reviews.text"].apply(lastWord)
dataset["endofreview"].head()
```

```python
#sending the word arrays to a .out file
wordsTxt = dataset["review.words"].values.flatten()
wordsArr = itertools.chain(wordsTxt)
temp=list(wordsArr)
arrOut=np.concatenate(temp, axis=0)
np.savetxt('AmazonPreprocessed.out', arrOut, delimiter=" ", fmt="%s",encoding='utf8')
```

```python
#sending the words arrays to a .txt file
outFile= open("AmazonPreprocessed.txt","w+",encoding='utf8')
outFile.write(" ".join(arrOut))
outFile.close()
```
